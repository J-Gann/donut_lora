```python
import json
```

```python
with open("documents/extraction.json", "r") as f:
    extraction = json.load(f)

with open("documents/german_pdf_files/mapping.json", "r") as f:
    pdf_mapping = json.load(f)

with open("documents/german_img_files/mapping.json", "r") as f:
    img_mapping = json.load(f)
```

```python
from urllib.parse import urlparse
import os
import shutil
```

```python
PDF_DIR = "documents/german_pdf_files"
IMG_DIR = "documents/german_img_files"
```

```python
pdf_mapping_new = {}
img_mapping_new = {}
extraction_new = {}
```

```python
import urllib.parse
```

```python
import unicodedata
```

```python
for url in extraction.keys():
    parsed_uri = urlparse(url)
    host = parsed_uri.hostname.replace(".", "_")

    decoded_str = urllib.parse.unquote(url)
    parsed_url = urlparse(decoded_str)
    file_name = parsed_url.hostname + "_" + os.path.basename(parsed_url.path)

    pdf_path = os.path.join(PDF_DIR, file_name)
    img_path = os.path.join(IMG_DIR, file_name.replace(".pdf", ".jpg"))

    new_pdf_path = os.path.join(host, file_name)
    new_img_path = os.path.join(host, file_name.replace(".pdf", ".jpg"))

    new_pdf_path = ''.join(char for char in new_pdf_path if char.isprintable())
    new_img_path = ''.join(char for char in new_img_path if char.isprintable())

    #os.makedirs(os.path.dirname(new_pdf_path), exist_ok=True)
    #os.makedirs(os.path.dirname(new_img_path), exist_ok=True)

    #shutil.copy(unicodedata.normalize("NFC", pdf_path), new_pdf_path)
    #shutil.copy(unicodedata.normalize("NFC", img_path), new_img_path)


    pdf_mapping_new[url] = new_pdf_path    
    img_mapping_new[url] = new_img_path

    extraction_new[url] = extraction[url]

with open("documents/german_pdf_files/mapping.json", "w", encoding="utf-8") as f:
    json.dump(pdf_mapping_new, f)

with open("documents/german_img_files/mapping.json", "w", encoding="utf-8") as f:
    json.dump(img_mapping_new, f)

#with open("documents/extraction_new.json", "w") as f:
#    json.dump(extraction_new, f)

```

```python
# delete excess files
for f in pdf_files:
    os.remove(os.path.join(PDF_DIR, f))

for f in img_files:
    os.remove(os.path.join(IMG_DIR, f))
```
